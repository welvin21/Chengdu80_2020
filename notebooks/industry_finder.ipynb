```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
transaction_data = pd.read_csv('../datasets/transaction_data.tsv', sep='\t')
```

```python
info = yf.Ticker('FB').info
print(info['sector'], info['industry'])
```

```python
relations = {}
i=1
for ticker in transaction_data['TICKER'].unique():
    print(f"{i}.{ticker}")
    i += 1
    try:
        info = yf.Ticker(ticker).info
        relations[ticker] = info['sector']
    except:
        relations[ticker] = ""
```

```python
cars = {'TICKER': list(relations.keys()),
        'Industry': list(relations.values())
        }

df = pd.DataFrame(cars, columns= ['TICKER', 'Industry'])
df.to_csv (r'../datasets/industry_relations.csv', index = False, header=True)
```
